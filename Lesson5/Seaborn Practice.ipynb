```python
# Seaborn is built upon Matplotlib 
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

# this turns off warning messages
import warnings
warnings.filterwarnings('ignore')
```

```python
sns.set_style('whitegrid') # visual style of Seaborn plots to have a white grid background. 
```

```python
# Let us use the Titanic dataset 
titanic = sns.load_dataset('titanic') # built in dataset
```

```python
# Reminder of the variables
titanic.head()
```

#### Kernel Density Estimate (KDE) Plot is used to plot the distribution of continuous data. The plot is a non-parametric way to estimate the probability density function of a continuous random variable.

```python
sns.kdeplot(x = 'age' , data = titanic , color = 'blue')
```

#### Let us look at some distributions. The sns.distplot() function from the Seaborn library creates a histogram with 30 bins for the 'fare' column in the Titanic dataset and set the color of the bars to red. The kde=False argument disables the Kernel Density Estimation plot, so you'll only see the histogram. It is set to False by default. However, if you wish to plot a KDE graph on top of the bars, you can set it to True.

```python
sns.distplot(titanic['fare'],bins=30,kde=False,color='red')
```

#### Let us create a box plot using Seaborn's boxplot function. The box plot will visualize the distribution of the 'age' variable across different 'class' categories from the 'titanic' dataset, using the 'rainbow' color palette

```python
sns.boxplot(x='class',y='age',data=titanic,palette='rainbow' )
```

#### Let us create a swarm plot which  is a categorical scatter plot that shows individual data points along the y-axis and groups them based on the categorical variable specified on the x-axis.

```python
sns.swarmplot(x='survived',y='age', data=titanic, palette='rainbow')
```
